```python
import pandas as pd

df = pd.read_csv("../jobs.csv")

df.head()
```

```python
df.shape  #Check Dataset Shape
```

```python
df.columns #checking colums names
```

```python
df.info() #Shows:Data types, Missing values ,Total records
```

```python
df.isnull().sum() #check missing values
```

```python
df.duplicated().sum() #Check Duplicate Records
```

```python
df = df.drop_duplicates() #if duplicate is there
```

```python
df.describe(include="all") #statistics part - gives summary of data
```

```python
#common companies - question: which company appear most?
df["Company"].value_counts().head(10)
```

```python
df["Location"].value_counts().head(10) 
#question: which location have most jobs?
```

```python
#visualization starts
import matplotlib.pyplot as plt
```

```python
# top locations
top_locations = df["Location"].value_counts().head(10)

top_locations.plot(kind="barh")

plt.title("Top Job Locations")
plt.xlabel("Location")
plt.ylabel("Count")

plt.show()
```

```python
#using seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.countplot(
    y=df["Location"],
    order=df["Location"].value_counts().head(10).index
)

plt.title("Top Job Locations")
plt.show()
```

```python
df.to_csv("cleaned_jobs.csv", index=False)
print("data saved")
# save clean data
```
