```python
import requests # download webpage
from bs4 import BeautifulSoup # read webpage
import pandas as pd # save data
```

```python
url = "https://realpython.github.io/fake-jobs/" # used this website - designed for scrapping, no login required, clean html.
```

```python
response = requests.get(url) # download webpage
```

```python
soup = BeautifulSoup(response.text, "lxml") #parse html
```

```python
# Find all job cards
jobs = soup.find_all("div", class_="card-content")
```

```python
# Empty list to store job data
job_data = []
```

```python
# Loop through each job card
for job in jobs:
    title = job.find("h2", class_="title").text.strip()
    company = job.find("h3", class_="company").text.strip()
    location = job.find("p", class_="location").text.strip()

    job_data.append({
        "Title": title,
        "Company": company,
        "Location": location,
        "Source": "Fake Jobs Website"
    })
```

```python
# Convert to DataFrame
df = pd.DataFrame(job_data)
```

```python
# Show first 5 rows
print(df.head())
```
