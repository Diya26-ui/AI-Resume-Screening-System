```python
import pdfplumber

with pdfplumber.open("../resumes/resume2.pdf") as pdf:
    text = ""

    for page in pdf.pages:
        text += page.extract_text()

print(text[:1000])
#read a resume pdf- download your resume in pdf form
```

```python
print(text) #checks extracted texts
```

```python
import re

email = re.findall(
    r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
    text
)

print(email)
#extracting email
```

```python
phone = re.findall(
    r"\+?\d[\d\s\-]{8,15}",
    text
)

print(phone)
#phone number
```

```python
skills_list = [
    "python",
    "sql",
    "aws",
    "excel",
    "tensorflow",
    "pandas",
    "numpy",
    "machine learning"
]
```

```python
found_skills = []

for skill in skills_list:
    if skill.lower() in text.lower():
        found_skills.append(skill)

print(found_skills)
#checking skills
```
